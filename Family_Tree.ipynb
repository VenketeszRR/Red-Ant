```python
n=int(input("Num of Relationship "))

family=dict()
me = input("Enter your name ")
family[me]=dict()

print("name1 name2 relationship ")
i=0
while i<n:
  text=input().split()
  if text[0] not in family:
    print(text[0]+" doesn't exist in family tree")
    continue
  if text[2] not in family[text[0]]:
    family[text[0]][text[2]]=[]
  family[text[0]][text[2]].append(text[1])
  if text[1] not in family:
    family[text[1]]=dict()
  i=i+1

print(family)
```

```python
from graphviz import Digraph

#dot = Digraph(comment = 'Ancestry', graph_attr = {'splines':'ortho'})

dot = Digraph(comment = 'Family Tree', graph_attr = {'splines':'ortho'})
for person in family:
  for relation in family[person]:
    for temp in family[person][relation]:
      dot.edge(person,temp,relation)

dot
```
